# Day 3 — numpy·pandas (데이터 불러오기·필터·기초통계)
- numpy: 벡터화 연산, 내적 = np.dot

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])   # 어제 화살표로 본 '벡터'가 코드로는 이것

print(arr + 10)     # [11 12 13 14 15]
print(arr * 2)      # [2 4 6 8 10]
print(arr.mean())   # 3.0
```

```python
big_list = list(range(1_000_000))
big_arr  = np.arange(1_000_000)

%timeit [x * 2 for x in big_list]   # 리스트 방식
%timeit big_arr * 2                  # 배열 방식 → 수십 배 빠름
```

%timeit 결과 읽는 법이에요. 하나씩 뜯어볼게요.
35 ms ± 730 µs per loop — 이게 핵심 결과입니다.

35 ms = 코드 한 번 실행하는 데 평균 35밀리초(0.035초) 걸렸다.
± 730 µs = 실행할 때마다 시간이 조금씩 들쭉날쭉한 정도(표준편차)가 730마이크로초(0.00073초). 즉 ±0.73ms 정도 흔들렸다는 뜻.
per loop = "한 번 실행당" 시간이라는 의미.

단위가 ms랑 µs로 섞여서 헷갈리실 텐데, 정리하면 1초 = 1,000ms = 1,000,000µs예요. 그러니까 730µs는 0.73ms고, "평균 35ms인데 오차가 0.73ms 정도니까 꽤 안정적으로 35ms 근처"라는 얘기입니다.
7 runs, 10 loops each — 이건 어떻게 쟀는지 측정 방법이에요.
%timeit은 한 번만 재면 우연히 느리거나 빠를 수 있으니 여러 번 재서 신뢰도를 높입니다. 코드가 워낙 빨라서 1회 시간을 정확히 못 재거든요. 그래서:

10 loops = 코드를 10번 연속 돌려서 그 시간을 잰 걸 한 묶음으로 하고,
7 runs = 그 묶음을 7세트 반복한 뒤,
7세트 중 가장 빠른 세트를 기준으로 "한 번당 평균"을 계산해서 보여줍니다.

loops 개수(10)는 %timeit이 코드 속도를 보고 자동으로 정해요. 빠른 코드면 loops를 수천 번으로 늘리고, 느린 코드면 몇 번으로 줄입니다.
그래서 선생님이 보신 결과는 — "이 코드는 한 번 실행에 약 35ms 걸리고, 측정이 안정적이다" 로 읽으시면 됩니다.
리스트 방식: 35 ms = 35,000 µs
배열(numpy) 방식: 445 µs

배수로 계산하면 35,000 ÷ 445 ≈ 약 79배. numpy가 리스트보다 79배쯤 빠르다는 뜻이에요. 같은 100만 개 곱하기를 한 건데 말이죠.
여기서 학생들한테 보여줄 포인트가 하나 더 숨어 있어요. 괄호 안 loops each 숫자가 다르다는 거예요.

리스트: 10 loops each
배열: 1000 loops each

%timeit은 코드가 빠를수록 더 많이 반복해서 잽니다(빨라서 한두 번으론 시간 측정이 부정확하니까요). 그래서 느린 리스트는 10번만, 빠른 배열은 1000번씩 돌린 거예요. 이 숫자 차이 자체가 "배열이 훨씬 빠르다"는 또 다른 증거인 셈이죠. 측정 결과(79배)와 측정 방식(10 vs 1000 loops)이 같은 이야기를 하고 있는 겁니다.

 "같은 계산도 자료구조를 잘 고르면 수십 배 빨라진다." 이게 사실 고2 자료구조 단원(§3-2)에서 "왜 numpy 배열이, 왜 적절한 자료구조가 대규모 AI에서 중요한가"로 이어지는 복선이에요


```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.dot(a, b))   # 32
```

한 가지 주의: Colab은 런타임이 끊기면(오래 방치하거나 닫았다 열면) 설치한 폰트가 날아가요. 그러면 위 셀을 다시 한 번 실행하면 됩니다. 그래서 이 폰트 설정 셀을 노트북 맨 위쪽에 둬서, 매번 제일 먼저 실행하는 습관을 들이시면 편해요.

핵심은 **"노트북 셀 실행"이 아니라 "런타임(세션) 단위"**라는 거예요. 한 번 설정하면 그 세션이 살아있는 동안에는 계속 유지됩니다. 그러니까:

같은 세션 안에서 그래프를 10개 그리든 100개 그리든 → 설정은 맨 처음 한 번만.
런타임이 끊기면 (노트북 닫았다 다시 열기, 오래 방치해서 연결 해제, 런타임 → 다시 시작) → 폰트가 날아가서 다시 한 번 해줘야 함.

즉 "matplotlib 그릴 때마다"가 아니라 "노트북 새로 열 때마다 한 번" 이에요. 하루 작업하면서 보통 1번, 많아야 2~3번이죠.
왜 이렇게 되냐면, Colab은 내 컴퓨터가 아니라 구글이 빌려주는 임시 가상머신이라 그래요. 세션 끝나면 깨끗이 초기화돼서, 설치했던 폰트도 같이 사라집니다. 이건 폰트만이 아니라 !pip install로 깐 라이브러리, 업로드한 파일도 다 마찬가지예요. (어제 말씀드린 "변수는 다시 실행하면 복구" 와 같은 원리예요.)
그래서 현실적인 운영법은 이거예요. 폰트 설정 코드를 노트북 맨 위 첫 번째 셀에 고정해두세요. 그리고 노트북 열면 그 셀부터 실행하는 습관. 더 편하게 하려면 런타임 → 모두 실행을 누르면 첫 셀(폰트 설정)부터 알아서 쭉 실행되니까 신경 쓸 일이 없어요.
혹시 "진짜 한 번도 신경 쓰기 싫다" 하시면 두 가지 방법이 더 있어요. 하나는 그래프 라벨을 그냥 영어로 쓰는 것 (xlabel("total_bill") 식으로) — 많은 실무자들이 실제로 이렇게 합니다.

markdown# Day 5 — ML이란 / 지도 vs 비지도
- 규칙 기반 프로그래밍: 사람이 규칙을 직접 코딩 (if 안경이면...)
- 학습 기반(ML): 데이터에 정답을 붙여주면 규칙을 스스로 익힘
- 방금 만든 분류기 = 지도학습 (라벨링 → 학습 → 예측)

Teachable Machine 모델 링크: (여기에 내보낸 링크 붙이기)
이 "규칙 기반 vs 학습 기반" 3줄이 사실 Day 1 지형도의 핵심 메시지 — "프로그래밍은 규칙을 직접 짜기, ML은 규칙을 데이터에서 찾게 하기" — 와 똑같아요. 1주차 내내 깔아온 복선이 오늘 학생 손으로 직접 만든 분류기에서 완성되는 거죠.
수업 연결. 이게 고1·고2 공통 흥미 유발 도입 차시예요. 학기 초에 이걸로 시작하면 "AI 수업 재밌겠다"는 첫인상이 잡혀요. 학생들이 자기 얼굴·물건으로 직접 분류기를 만들고, 그게 틀리는 것도 보면서 "데이터를 잘 줘야 잘 맞히는구나"까지 자연스럽게 체감합니다. 라벨을 일부러 엉터리로 줘서 틀리게 해보는 것도 좋은 활동이에요(데이터 품질 감각 → 나중에 편향 토론으로 연결).

이 경우 같은 "모델 내보내기" 창에서 이렇게 하세요:

Tensorflow.js 탭이 선택된 상태에서 (기본값이에요),
위쪽에 "모델 업로드(공유 가능한 링크)" 와 "다운로드" 중 → 이번엔 "모델 업로드" 선택.
"모델 업로드" 버튼을 누르면 구글 서버에 올라가고, 잠시 뒤 공유 가능한 링크가 하나 생성됩니다.
그 링크를 복사해두세요. → 이게 학생들에게 줄 주소예요.

이 링크의 좋은 점은, 학생들이 아무것도 설치할 필요 없이 브라우저에서 링크만 열면 바로 그 모델을 자기 웹캠으로 테스트해볼 수 있다는 거예요. 수업에서 "이 링크 열어봐" 한 줄이면 끝납니다.
다만 알아두실 점 두 가지가 있어요.
첫째, **이 링크는 공개(누구나 접근 가능)**예요. 주소를 아는 사람은 볼 수 있으니, 물건 모델이라 상관없지만 "완전 공개"라는 것만 기억하세요. 그래서 얼굴 모델은 안 올리고 이건 올리는 게 맞는 거고요.
둘째, 모델을 수정하면 링크를 다시 만들어야 해요. 나중에 샘플을 더 추가하거나 클래스를 바꾸면, "모델 업로드"를 다시 눌러서 새 링크를 받아야 최신 버전이 반영됩니다. 기존 링크는 옛날 모델 그대로예요.
그리고 권하고 싶은 건, 링크 업로드 + 다운로드 둘 다 해두는 거예요.
